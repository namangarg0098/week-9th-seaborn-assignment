```python
Seaborn can be used to plot many plots. Some of them are:

Scatterplot : Scatterplot are used to visualise relationship between any two variables. The point on scatterplot represents the data point value of both the variables

Barplot : Barplot is used to represent categorical data with help of rectangular bars, where height of each bar represents the aggregated value.

Relplot : Relplot is used to represent the relationship between variables. It is used when we want to explore interactions, patterns, and interconnections between mulitple varibles

catplot : catplot is used to create categorical plots. It helps to create various types of plots for visualising relations between categorical data.

Jointplot : jointplot is used to create joint plots, which combine two different plots to represent relationship of two variables.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a pivot table
pivot_table = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5)
plt.title('Passenger Count by Month and Year')
plt.show()
```
